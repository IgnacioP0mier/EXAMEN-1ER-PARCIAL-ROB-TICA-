```python
import sympy as sp

theta = sp.Symbol('q1')
d     = 0
a     = sp.Symbol('l1')
alpha = 0


T1 = sp.Matrix([
    [sp.cos(theta), -sp.sin(theta)*sp.cos(alpha),  sp.sin(theta)*sp.sin(alpha), a*sp.cos(theta)],
    [sp.sin(theta),  sp.cos(theta)*sp.cos(alpha), -sp.cos(theta)*sp.sin(alpha), a*sp.sin(theta)],
    [0,              sp.sin(alpha),               sp.cos(alpha),                d],
    [0,              0,                           0,                            1]
])

sp.pprint(T1, use_unicode=True)
```

```python
import sympy as sp

theta = sp.Symbol('q2')
d     = 0
a     = sp.Symbol('l2')
alpha = 0


T2 = sp.Matrix([
    [sp.cos(theta), -sp.sin(theta)*sp.cos(alpha),  sp.sin(theta)*sp.sin(alpha), a*sp.cos(theta)],
    [sp.sin(theta),  sp.cos(theta)*sp.cos(alpha), -sp.cos(theta)*sp.sin(alpha), a*sp.sin(theta)],
    [0,              sp.sin(alpha),               sp.cos(alpha),                d],
    [0,              0,                           0,                            1]
])

sp.pprint(T2, use_unicode=True)
```

```python
import sympy as sp

theta = sp.Symbol('q3')
d     = 0
a     = sp.Symbol('l3')
alpha = 0


T3 = sp.Matrix([
    [sp.cos(theta), -sp.sin(theta)*sp.cos(alpha),  sp.sin(theta)*sp.sin(alpha), a*sp.cos(theta)],
    [sp.sin(theta),  sp.cos(theta)*sp.cos(alpha), -sp.cos(theta)*sp.sin(alpha), a*sp.sin(theta)],
    [0,              sp.sin(alpha),               sp.cos(alpha),                d],
    [0,              0,                           0,                            1]
])

sp.pprint(T3, use_unicode=True)
```

```python
import sympy as sp

theta = sp.Symbol('q4')
d     = 0
a     = sp.Symbol('l4')
alpha = 0


T4 = sp.Matrix([
    [sp.cos(theta), -sp.sin(theta)*sp.cos(alpha),  sp.sin(theta)*sp.sin(alpha), a*sp.cos(theta)],
    [sp.sin(theta),  sp.cos(theta)*sp.cos(alpha), -sp.cos(theta)*sp.sin(alpha), a*sp.sin(theta)],
    [0,              sp.sin(alpha),               sp.cos(alpha),                d],
    [0,              0,                           0,                            1]
])

sp.pprint(T4, use_unicode=True)
```

```python
T = T1 @ T2 @ T3 @ T4
print(sp.sstr(T))

```

```python
T_final = sp.simplify(T)
print(sp.sstr(T_final))
```

```python
import sympy as sp

valores = {
    sp.Symbol('q1'): 0,
    sp.Symbol('q2'): 0,
    sp.Symbol('q3'): 0,
    sp.Symbol('q4'): 0,
    
    sp.Symbol('l1'): 2,
    sp.Symbol('l2'): 1.5,
    sp.Symbol('l3'): 0.8,
    sp.Symbol('l4'): 0
}

Px = T_final[0,3].evalf(subs=valores)
Py = T_final[1,3].evalf(subs=valores)
Pz = T_final[2,3].evalf(subs=valores)

Px, Py, Pz
```

# **Matriz Jacobiana (inversa)**

```python
import sympy as sp

x = sp.simplify(T_final[0,3])
y = sp.simplify(T_final[1,3])
z = sp.simplify(T_final[2,3])

pos = sp.Matrix([x, y, z])

simbolos = {str(s): s for s in T_final.free_symbols}

q1_ = simbolos['q1']
q2_ = simbolos['q2']
q3_ = simbolos['q3']
q4_ = simbolos['q4']

Jv = sp.simplify(pos.jacobian([q1_, q2_, q3_, q4_]))

print("Vector de posiciГіn:")
sp.pprint(pos, use_unicode=True)

print("\nJacobiana lineal Jv:")
sp.pprint(Jv, use_unicode=True)
```
